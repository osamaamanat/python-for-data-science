```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_data.csv")
data.head()
```

```python
data["cast"] = data["cast"].replace(np.NaN, "None")
data.head()
```

```python
def num_of_cast(arr):
    arr1 = arr.split(", ")
    return len(arr1)

data["num_of_cast"]= data.loc[:,"cast"].apply(num_of_cast)
data.head()
```
